```python
#importing libraries for EDA 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#check for nulls and deal with NaNs & update data types & feature engineering

#reopen_time delete and 1 and 0
inc["Reopened"] = inc["Reopen_Time"].apply(lambda x: 1 if(pd.notnull(x)) else 0)

# drop unnecessary columns
inc.drop(['No_of_Related_Changes', 'Related_Change', 'Reopen_Time', 'KB_number'], inplace = True, axis=1)


plt.figure(figsize=(25,6))
sns.heatmap(inc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
#impact/urgency -> numeric
#set numeric values + median of NaNs
def number_parse(column):
    inc[column] = pd.to_numeric(inc[column], errors='coerce')
    inc[column].fillna(inc[column].median(), inplace = True)

impact = "Impact"
urgency = "Urgency"
priority = "Priority"

number_parse(impact)
number_parse(urgency)
number_parse(priority)
```

```python
#open_time/resolve_time/close_time -> datetime 

open_time = 'Open_Time'
resolved_time = 'Resolved_Time'
close_time = 'Close_Time'

def datetime_parse(column):
    inc[column] = pd.to_datetime(inc[column], errors='coerce')

datetime_parse(open_time)
datetime_parse(resolved_time)
datetime_parse(close_time)

#handle_time_hrs -> timestamp
```

```python
#if resolved time is empty take close time
inc["Resolved_Time"].fillna(inc["Close_Time"], inplace=True)
```

```python
#drop nan's row from Closure_Code & No_of_Related_Interactions
#drop No_of_Related_Incidents

inc.drop("No_of_Related_Incidents", axis=1, inplace=True)
inc.dropna(how='any', inplace=True)
plt.figure(figsize=(25,6))
sns.heatmap(inc.isnull(),yticklabels=False,cbar=False,cmap='viridis')
```

```python
inc['Handle_Time_hrs']=inc['Handle_Time_hrs'].str.replace(',',':')
```

```python
inc["prio"] = inc["Impact"]*inc["Urgency"]*inc["Priority"]
inc.drop(["Impact", "Urgency", "Priority", "Alert_Status"], axis=1, inplace=True)
```

```python
#numerize category column --> map
inc["Category"] = inc["Category"].map({"incident": 1, "request for information":2, "complaint":3})
```

```python
#drop all WIP in STATus + drop status + copy of dataframe
inc_clean = inc.copy()
inc_clean = inc_clean[inc_clean.Status != "Work in progress"]
#inc_clean.drop("Categorry", axis=1, inplace = True)
inc_clean.head()
```

```python
#check for duplicates + drop further columns
inc_clean.Incident_ID.duplicated().any()
inc_clean.drop(["CI_Name", "WBS", "Status", "Related_Interaction"], axis=1, inplace=True)
inc_clean.head()
```

```python
# replace handle_time_hrs in seconds -> remove Open, resolved, closed time & handle_time_hrs
inc_clean["handle_time_sec"] = (inc_clean.Close_Time-inc_clean.Open_Time).astype('timedelta64[s]')
inc_clean.drop(["Open_Time", "Resolved_Time", "Close_Time", "Handle_Time_hrs"], axis=1, inplace = True)
```

```python
#enumerate categorical columns decided to do the labelEncoder not getdummies or one-hot-encoding even though there is a risk that ML-Algorithm give those categories a hirachical meaning...
from sklearn.preprocessing import LabelEncoder

cat = LabelEncoder()

inc_clean_ml["CI_Cat"] = cat.fit_transform(inc_clean_ml["CI_Cat"])
inc_clean_ml["CI_Subcat"] = cat.fit_transform(inc_clean_ml["CI_Subcat"])
inc_clean_ml["Closure_Code"] = cat.fit_transform(inc_clean_ml["Closure_Code"])

inc_clean_ml.head()

inc_clean_ml.drop("Incident_ID", inplace=True, axis=1)
```

```python
#percentage of outliers
for k,v in inc_clean_ml.items():
    q1 = v.quantile(0.25)
    q3 = v.quantile(0.75)
    iqr = q3 - q1
    v_col = v[(v <= q1 - 1.5 * iqr) | (v >= q3 + 1.5 * iqr)]
    perc = np.shape(v_col)[0] * 100.0 / np.shape(inc_clean_ml)[0]
    print("%s Outliers = %.2f%%" % (k, perc))
```
